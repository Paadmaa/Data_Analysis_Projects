## Data Analysis

### Top 5 most popular artist

```python
top_5_artists = df.groupby("artist").count().sort_values(by="song_title", ascending = False)["song_title"][:5]
top_5_artists
```

```python
top_5_artists.plot.barh()
plt.show()
```

### Top 5 loudest tracks

```python
top_5_loudest_tracks = df[["loudness","song_title"]].sort_values(by="loudness",ascending = False)
top_5_loudest_tracks
```

```python
top_5_loudest_tracks = df[["loudness","song_title"]].sort_values(by="loudness",ascending = True)[:5]
top_5_loudest_tracks
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="loudness",y="song_title",data = top_5_loudest_tracks)
plt.title("Top 5 loudest tracks")
plt.show()
```

### Artist with the most danceability song

```python
top_5_danceable_songs = df[["danceability", "song_title", "artist"]].sort_values(by="danceability",
                                                                                 ascending=False)[:5]
top_5_danceable_songs
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="danceability",y="artist",data=top_5_danceable_songs)
plt.title("Artist with the most Danceability Songs")
plt.show()
```

### Top 10 instrumentalness tracks

```python
df.columns
```

```python
top_5_instrumental_songs = df[["instrumentalness", "song_title", "artist"]].sort_values(by="instrumentalness",
                                                                                 ascending=False)[:5]
top_5_instrumental_songs
```

```python
plt.figure(figsize=(12,7))
plt.pie(x="instrumentalness",data = top_5_instrumental_songs, autopct="%1.2f%%",labels=top_5_instrumental_songs.song_title)
plt.title("Top 10 instrumentalness tracks")
plt.show()
```

### Multiple feature plots

```python
df = pd.read_csv("spotify_data.csv")
```

```python
interest_feature_cols = ["tempo","loudness","acousticness","danceability","duration_ms",
                        "energy","instrumentalness","liveness","speechiness","valence"]
```

```python
for feature_col in interest_feature_cols:
    pos_data = df[df["target"] == 1][feature_col]
    neg_data = df[df["target"] == 0][feature_col]
    
    plt.figure(figsize=(12,7))
    
    # Use histplot for overlaying histograms
    sns.histplot(pos_data, bins=30, label="Positive", color="green", kde=True, stat="density", alpha=0.5)
    sns.histplot(neg_data, bins=30, label="Negative", color="red", kde=True, stat="density", alpha=0.5)
    
    plt.legend(loc="upper right")
    plt.title(f"Positive And Negative Histogram Plot for {feature_col}")
    plt.show()
```
