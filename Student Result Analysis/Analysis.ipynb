```python
#Drop Unnamed Column
df = df.drop("Unnamed: 0",axis = 1)
```

```python
#Weekly Study hrs Column
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-oct","05-10")
```

```python
df.head()
```

```python
#Gender distribution
```

```python
plt.figure(figsize=(5,4))
sns.countplot(data = df, x = "Gender")
plt.show()
```

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#Conclusion - no. of Females students are more than of male students
```

```python
gp = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
```

```python
gp
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gp, annot = True)
plt.show()
```

```python
#Conclusion - Education of parents have a good impact on Students
```

```python
gp1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gp1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gp1, annot = True)
plt.show()
```

```python
#Conclusion - Marital Status of parents doesn't impact on students
```

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["group A", "group B", "group C", "group D", "group E"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"],
        groupD["EthnicGroup"], groupE["EthnicGroup"]]

plt.pie(mlist, labels = l, autopct="%1.2f%%")
plt.title("Distribution of Groups")
plt.show()
```
